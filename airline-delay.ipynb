# Delayed Rates by carrier name from 2022-2024

```python
carrier_performance_agg = delay_df.groupby('carrier_name').agg(totalflights = ('arr_flights', 'sum'),
                                                          delayedflights = ('arr_del15','sum'))
carrier_performance_agg     
```

```python
carrier_performance_agg['delayed_rates_percent'] = ((carrier_performance_agg['delayedflights']/carrier_performance_agg['totalflights'])*100).round(2)
carrier_performance_agg = carrier_performance_agg.sort_values(by = 'delayed_rates', ascending = False)
carrier_performance_agg = carrier_performance_agg.sort_values(by = ['totalflights','delayed_rates'], ascending = [False, False])
carrier_performance_agg
```

```python
carrier_performance_agg[carrier_performance_agg['delayed_rates_percent'] < 21.29].sort_values(by = 'delayed_rates_percent', ascending = True).head(5)
```

# Delay of flights per month by carrier

```python
delayed_months =(delay_df.groupby(['carrier_name', 'month'])[['arr_del15', 'arr_flights']].sum().reset_index())
delayed_months['percent_delayed_m'] = (delayed_months['arr_del15'] / delayed_months['arr_flights']).round(4)*100
delayed_months
```

# Delay of flights per year by carrier

```python
delayed_months =(delay_df.groupby(['carrier_name', 'year'])[['arr_del15', 'arr_flights']].sum().reset_index())
delayed_months['percent_delayed_y'] = (delayed_months['arr_del15'] / delayed_months['arr_flights']).round(4)*100
delayed_months
```
