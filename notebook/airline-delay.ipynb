# Delayed Rates by carrier name from 2022-2024

```python
carrier_performance_agg = delay_df.groupby('carrier_name').agg(totalflights = ('arr_flights', 'sum'),
                                                             delayedflights = ('arr_del15','sum'),
                                                              delayedmin = ('arr_delay','sum'))
carrier_performance_agg     
```

```python
carrier_performance_agg['delayed_rates_percent'] = ((carrier_performance_agg['delayedflights']/carrier_performance_agg['totalflights'])*100).round(2)
carrier_performance_agg['avg_min_overall'] = (carrier_performance_agg['delayedmin']/carrier_performance_agg['delayedflights'])
carrier_performance_agg = carrier_performance_agg.sort_values(by = 'delayed_rates_percent', ascending = False)
carrier_performance_agg = carrier_performance_agg.sort_values(by = ['totalflights','delayed_rates_percent'], ascending = [False, False])
```

```python
carrier_performance_agg = carrier_performance_agg.rename(columns = {'totalflights': 'Total Flights Overall',
                                                                    'carrier_name': 'Airline',
                                                                    'delayedflights': '# of Overall Delayed Flights',
                                                                    'delayed_rates_percent': 'Flights Delayed Overall (%)', 
                                                                    'avg_min_overall':'Overall Average Flight Delay (Minutes)',
                                                                   'delayedmin':'Overall Count of Minutes Delayed'})
carrier_performance_agg
```

# Delay of flights per month by carrier

```python
delayed_months =(delay_df.groupby(['carrier_name', 'month'])[['arr_del15', 'arr_flights', 'arr_delay']].sum().reset_index())
delayed_months['percent_delayed_m'] = (delayed_months['arr_del15'] / delayed_months['arr_flights']).round(4)*100
delayed_months = delayed_months[['percent_delayed_m','carrier_name', 'month','arr_del15', 'arr_flights', 'arr_delay']]
```

# Delayed flights per year by carrier

```python
delayed_year =(delay_df.groupby(['carrier_name', 'year'])[['arr_del15', 'arr_flights', 'arr_delay']].sum().reset_index())
delayed_year['percent_delayed_y'] = (delayed_year['arr_del15'] / delayed_year['arr_flights']).round(4)*100
delayed_year = delayed_year[['percent_delayed_y','carrier_name', 'year','arr_del15', 'arr_flights', 'arr_delay']]
delayed_year
```

# average minutes of flights by minute by carrier per month

```python
avg_min_months =(delay_df.groupby(['carrier_name', 'month'])[['arr_flights','arr_del15', 'arr_delay']].sum().reset_index())
avg_min_months['avg_min_m'] = (avg_min_months['arr_delay'] / avg_min_months['arr_del15'])
avg_min_months = avg_min_months[['avg_min_m', 'carrier_name','month', 'arr_flights','arr_del15', 'arr_delay']]
avg_min_months 
```

```python
avg_min_years =(delay_df.groupby(['carrier_name', 'year'])[['arr_flights','arr_del15', 'arr_delay']].sum().reset_index())
avg_min_years['avg_min_y'] = (avg_min_years['arr_delay'] / avg_min_years['arr_del15'])
avg_min_years = avg_min_years[['avg_min_y','carrier_name','year', 'arr_flights','arr_del15', 'arr_delay']]
avg_min_years
```
