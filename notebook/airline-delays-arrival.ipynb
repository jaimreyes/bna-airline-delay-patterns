```python
import pandas as pd
import numpy as np
import requests
import zipfile
from pathlib import Path
import pandas as pd
from io import BytesIO, StringIO
import time
import os
from datetime import datetime, date, time 
```

```python
folder_path = 'D:/Jai - NSS/bna-airline-delay-patterns/delay-zip-files'

# List to store all DataFrames
all_dfs = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(".zip"):
        zip_path = os.path.join(folder_path, file_name)
        
        with zipfile.ZipFile(zip_path, 'r') as zip_file:
            for inner_file in zip_file.namelist():
                # Check if the inner file is a CSV
                if inner_file.endswith(".csv"):
                    with zip_file.open(inner_file) as f:
                        # Read bytes and decode to string
                        data = f.read().decode('utf-8')
                        
                        # Load CSV into a DataFrame
                        df = pd.read_csv(StringIO(data))
                        all_dfs.append(df)

# Combine all DataFrames into one
combined_df = pd.concat(all_dfs, ignore_index=True)
print(combined_df.head())
#https://www.geeksforgeeks.org/python/working-zip-files-python/
##little side note 'utf-8' is how python reads bytes to string. If you do not add 'utf-8' it will read all the data raw. There are other like 'latin-1', 'iso-8859-1' ,'utf-16' If you try to decode a file with the wrong encoding, Python will raise a UnicodeDecodeError
```

#### There are two common ways to assign seasons to months:
####  1. Using a dictionary to map months to seasons
####  2. Creating a user-defined function
#### Here’s an example using a dictionary:
#### season_map = {1: "Winter", 2: "Winter",3: "Spring", 4: "Spring", 5: "Spring", 6: "Summer", 7: "Summer", 8: "Summer", 9: "Fall", 10: "Fall", 11: "Fall", 12: "Winter"}

#### The map() function can then be used to apply this mapping to a pandas Series. It applies a given function or mapping to each element of an iterable (like a list or Series) and returns a new mapped object.
#### For more details, see: https://www.geeksforgeeks.org/python/python-map-function/

```python
def season_map(season):
    if season <=2:
        return 'winter'
    elif 3 <= season <=5:
        return 'spring'
    elif 6 <= season <= 8:
        return 'summer'
    elif 9 <= season <= 11:
        return 'fall'
    else:
        return 'winter'


bna_df['season'] = bna_df['month'].map(season_map)
bna_df.sample(5)
```

# THIS IS FOR ARRIVAL ON TIME PERFORMANCE

```python
# Aggregate sums by carrier
carrier_performance_agg = bna_df.groupby('airline')[['flights',
                                                                'arr_del15',
                                                                'arr_delay_new', 
                                                                'dep_del15', 
                                                                'dep_delay_new',
                                                                'cancelled', 
                                                                'diverted', 
                                                                'carrier_delay_flight',
                                                                'weather_delay_flight',
                                                                'nas_delay_flight', 
                                                                'security_delay_flight', 
                                                                'late_aircraft_delay_flight']].sum().reset_index()

#Percentages
carrier_performance_agg['Arrival Flight Delays (%)'] = (
    carrier_performance_agg['arr_del15'] / carrier_performance_agg['flights'] * 100).round(2)

carrier_performance_agg['% cancelled'] = (
    carrier_performance_agg['cancelled'] / carrier_performance_agg['flights'] * 100).round(2)

carrier_performance_agg['% diverted'] = (
    carrier_performance_agg['diverted'] / carrier_performance_agg['flights'] * 100).round(2)

# Average delays (handle division by zero)
carrier_performance_agg['Avg. arrival delay (min)'] = np.where(
    carrier_performance_agg['arr_del15'] > 0,
    carrier_performance_agg['arr_delay_new'] / carrier_performance_agg['arr_del15'],0)

#Percent of flights delayed by cause
flights = carrier_performance_agg['flights'] 

carrier_performance_agg['% delayed due to carrier'] = (carrier_performance_agg['carrier_delay_flight'] / 
                                                       flights* 100).round(2)

carrier_performance_agg['% delayed due to weather'] = (carrier_performance_agg['weather_delay_flight'] / 
                                                       flights * 100).round(2)

carrier_performance_agg['% delayed due to NAS'] = (carrier_performance_agg['nas_delay_flight'] / 
                                                   flights * 100).round(2)

carrier_performance_agg['% delayed due to security'] = (carrier_performance_agg['security_delay_flight'] / 
                                                        flights * 100).round(2)

carrier_performance_agg['% delayed due to late aircraft'] = (carrier_performance_agg['late_aircraft_delay_flight'] / 
                                                             flights * 100).round(2)

#Column order
columns_order = ['airline',
                    'flights',
                    'arr_del15',
                    'Arrival Flight Delays (%)',
                    'Avg. arrival delay (min)',
                    'cancelled',
                    '% cancelled',
                    '% diverted',
                    '% delayed due to carrier',
                    '% delayed due to weather',
                    '% delayed due to NAS',
                    '% delayed due to security',
                    '% delayed due to late aircraft']

carrier_performance_agg = carrier_performance_agg[columns_order]

#Sort by arrival and departure delays
carrier_performance_agg = carrier_performance_agg.sort_values(by='Arrival Flight Delays (%)', ascending= False).reset_index(drop = True)
carrier_performance_agg
```
