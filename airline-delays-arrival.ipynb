```python
import pandas as pd
import numpy as np
import requests
import zipfile
from pathlib import Path
import pandas as pd
from io import BytesIO, StringIO
import time
import os
from datetime import datetime, date, time 
```

```python
folder_path = "D:\\Jai - NSS\\bna-airline-delay-patterns\\bna-delay-cause"

# List to store all DataFrames
all_dfs = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(".zip"):
        zip_path = os.path.join(folder_path, file_name)
        
        with zipfile.ZipFile(zip_path, 'r') as zip_file:
            for inner_file in zip_file.namelist():
                # Check if the inner file is a CSV
                if inner_file.endswith(".csv"):
                    with zip_file.open(inner_file) as f:
                        # Read bytes and decode to string
                        data = f.read().decode('utf-8')
                        
                        # Load CSV into a DataFrame
                        df = pd.read_csv(StringIO(data))
                        all_dfs.append(df)

# Combine all DataFrames into one
combined_df = pd.concat(all_dfs, ignore_index=True)
print(combined_df.head())
#https://www.geeksforgeeks.org/python/working-zip-files-python/
##little side note 'utf-8' is how python reads bytes to string. If you do not add 'utf-8' it will read all the data raw. There are other like 'latin-1', 'iso-8859-1' ,'utf-16' If you try to decode a file with the wrong encoding, Python will raise a UnicodeDecodeError
```

#### There are two common ways to assign seasons to months:
     1. Using a dictionary to map months to seasons
     2. Creating a user-defined function
#### Here’s an example using a dictionary:
season_map = {1: "Winter", 2: "Winter",3: "Spring", 4: "Spring", 5: "Spring", 6: "Summer", 7: "Summer", 8: "Summer", 9: "Fall", 10: "Fall", 11: "Fall", 12: "Winter"}
##### The map() function can then be used to apply this mapping to a pandas Series. It applies a given function or mapping to each element of an iterable (like a list or Series) and returns a new mapped object.

##### For more details, see: https://www.geeksforgeeks.org/python/python-map-function/

```python
def season_map(season):
    if season <=2:
        return 'winter'
    elif 3 <= season <=5:
        return 'spring'
    elif 6 <= season <= 8:
        return 'summer'
    elif 9 <= season <= 11:
        return 'fall'
    else:
        return 'winter'


bna_df['season'] = bna_df['month'].map(season_map)
bna_df.sample(5)
```

# THIS IS  ON TIME PERFORMANCE

```python
carrier_performance_agg = bna_df.groupby('branded_code_share').agg(totalflights = ('flights', 'sum'),
                                                             arr_delayedflights = ('arr_del15','sum'),
                                                             arr_delayedmin = ('arr_delay_new','sum'),
                                                             dep_delayedmin = ('dep_delay_new','sum'),
                                                             dep_delayedflights = ('dep_del15','sum'))     
carrier_performance_agg     
```

```python
carrier_performance_agg['Arrival Flight Delays (%)'] = ((carrier_performance_agg['arr_delayedflights']/carrier_performance_agg['totalflights'])*100).round(2)
carrier_performance_agg['Departure Flight Delays(%)'] = ((carrier_performance_agg['dep_delayedflights']/carrier_performance_agg['totalflights'])*100).round(2)
carrier_performance_agg['Avg. arrival delay (min)'] = (carrier_performance_agg['arr_delayedmin']/carrier_performance_agg['arr_delayedflights'])
carrier_performance_agg['Avg. departure delay(min)'] = (carrier_performance_agg['dep_delayedmin']/carrier_performance_agg['dep_delayedflights'])
carrier_performance_agg = carrier_performance_agg.sort_values(by = ['Arrival Flight Delays (%)','Departure Flight Delays(%)'], ascending = [False, False])
```

```python
carrier_performance_agg = carrier_performance_agg.rename({'totalflights': '# of total flights', 
                                                          'arr_delayedflights': '# of total arrival flights',
                                                          'arr_delayedmin': '# total arrival delayed min',
                                                          'dep_delayedmin': '# total departure delayed min',
                                                          'dep_delayedflights': '# of total departure flights'})
carrier_performance_agg 
```

# Overall delay rate

```python
avg_arr_delay_grouped = bna_df.groupby('branded_code_share')['arr_delay_new'].mean().reset_index()
avg_dep_delay_grouped = bna_df.groupby('branded_code_share')['dep_delay_new'].mean().reset_index()
avg_delay_grouped = pd.merge(avg_arr_delay_grouped, avg_dep_delay_grouped, on = 'branded_code_share')
```

```python
delay_summary = (
    bna_df.groupby(['branded_code_share', 'season', 'time_of_day'])
    .agg(total_flights=('flights', 'sum'),
        delayed_arrivals=('arr_del15', 'sum'),
        delayed_departures=('dep_del15', 'sum'))
    .reset_index())


delay_summary['arrival_delay_rate (%)'] = (delay_summary['delayed_arrivals'] / delay_summary['total_flights'] * 100).round(2)
delay_summary['departure_delay_rate (%)'] = (delay_summary['delayed_departures'] / delay_summary['total_flights'] * 100).round(2)
delay_summary['overall_delay_rate (%)'] = ((delay_summary['delayed_arrivals'] + delay_summary['delayed_departures']) / (2*delay_summary['total_flights']) * 100).round(2)
```

```python
# df1: Monthly percentages of delayed flights
df1 = (bna_df.groupby(['branded_code_share'])
       [['arr_del15', 'dep_del15', 'flights']]
       .sum()
       .reset_index())
df1['arrival flights monthly (%)'] = (df1['arr_del15'] / df1['flights']).round(4) * 100
df1['departure flights monthly (%)'] = (df1['dep_del15'] / df1['flights']).round(4) * 100
df1 = df1[['branded_code_share','arrival flights monthly (%)','departure flights monthly (%)']]

# df2: Yearly percentages of delayed flights
df2 = (bna_df.groupby(['branded_code_share'])
       [['arr_del15','dep_del15','flights']]
       .sum()
       .reset_index())
df2['arrival flights yearly (%)'] = (df2['arr_del15'] / df2['flights']).round(4) * 100
df2['departure flights yearly (%)'] = (df2['dep_del15'] / df2['flights']).round(4) * 100
df2 = df2[['branded_code_share','arrival flights yearly (%)','departure flights yearly (%)']]

# df3: Yearly average minutes delayed
df3 = (bna_df.groupby(['branded_code_share'])
       [['arr_delay_new','dep_delay_new','arr_del15','dep_del15']]
       .sum()
       .reset_index())
df3['Avg. min delayed for arrival flights (year)'] = df3['arr_delay_new'] / df3['arr_del15']
df3['Avg. min delayed for departure flights (year)'] = df3['dep_delay_new'] / df3['dep_del15']
df3 = df3[['branded_code_share','Avg. min delayed for arrival flights (year)','Avg. min delayed for departure flights (year)']]

# df4: Monthly average minutes delayed
df4 = (bna_df.groupby(['branded_code_share'])
       [['arr_delay_new','dep_delay_new','arr_del15','dep_del15']]
       .sum()
       .reset_index())
df4['Avg. min delayed for arrival flights (month)'] = df4['arr_delay_new'] / df4['arr_del15']
df4['Avg. min delayed for departure flights (month)'] = df4['dep_delay_new'] / df4['dep_del15']
df4 = df4[['branded_code_share','Avg. min delayed for arrival flights (month)','Avg. min delayed for departure flights (month)']]

# Merge all summaries into a single DataFrame
week_df = df1.merge(df2, on='branded_code_share', how='outer') \
             .merge(df3, on='branded_code_share', how='outer') \
             .merge(df4, on='branded_code_share', how='outer')
```

```python
# week_type_df1: Average minutes delayed by week_type, with year and month
week_type_df1 = (bna_df.groupby(['branded_code_share', 'week_type', 'year', 'month'])
                 [['arr_delay_new','dep_delay_new','arr_del15','dep_del15']]
                 .sum()
                 .reset_index())
week_type_df1['Avg. min delayed for arrival flights (week_type)'] = week_type_df1['arr_delay_new'] / week_type_df1['arr_del15']
week_type_df1['Avg. min delayed for departure flights (week_type)'] = week_type_df1['dep_delay_new'] / week_type_df1['dep_del15']
week_type_df1 = week_type_df1[['branded_code_share','week_type','year','month',
                               'Avg. min delayed for arrival flights (week_type)',
                               'Avg. min delayed for departure flights (week_type)']]

# week_type_df2: Percent of flights delayed by week_type
week_type_df2 = (bna_df.groupby(['branded_code_share','week_type','year','month'])
                 [['arr_del15','dep_del15','flights']]
                 .sum()
                 .reset_index())
week_type_df2['arrival flights week_type (%)'] = (week_type_df2['arr_del15'] / week_type_df2['flights']).round(4) * 100
week_type_df2['departure flights week_type (%)'] = (week_type_df2['dep_del15'] / week_type_df2['flights']).round(4) * 100
week_type_df2 = week_type_df2[['branded_code_share','week_type','year','month',
                               'arrival flights week_type (%)','departure flights week_type (%)']]

# week_type_df3: Yearly averages
week_type_df3 = (bna_df.groupby(['branded_code_share','year'])
                 [['arr_delay_new','dep_delay_new','arr_del15','dep_del15']]
                 .sum()
                 .reset_index())
week_type_df3['Avg. min delayed for arrival flights (year)'] = week_type_df3['arr_delay_new'] / week_type_df3['arr_del15']
week_type_df3['Avg. min delayed for departure flights (year)'] = week_type_df3['dep_delay_new'] / week_type_df3['dep_del15']
week_type_df3 = week_type_df3[['branded_code_share','year',
                               'Avg. min delayed for arrival flights (year)',
                               'Avg. min delayed for departure flights (year)']]

# Merge all metrics together
week_type_df = week_type_df1.merge(week_type_df2, on=['branded_code_share','week_type','year','month'], how='outer') \
                            .merge(week_type_df3, on=['branded_code_share','year'], how='outer')

# Sort by year, month, and week_type
week_type_df = week_type_df.sort_values(by=['year','month','week_type'])
week_type_df
```
